```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict

# Read the HTML file
with open("vegconomistlocal.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Extract article URLs and company tags
article_list_local_html = [link.a["href"] for link in soup.find_all("h2", class_="entry-title")]

# Dictionary to store article count per company
articles_media_per_company = defaultdict(int)

# Loop through each article and extract tagged companies
for article_url in article_list_local_html:
    article_response = requests.get(article_url)
    article_soup = BeautifulSoup(article_response.text, "html.parser")

    # Find company links underneath the article title
    company_links = article_soup.find_all("a", class_="post-term-link")

    # Extract company names and update dictionary
    for company in company_links:
        company_name = company.text.strip().lower()  # Normalize case
        articles_media_per_company[company_name] += 1

# Sort dictionary alphabetically by company name
articles_media_per_company = dict(sorted(articles_media_per_company.items()))

# Print the sorted dictionary
print(articles_media_per_company)
```

```python
# Count occurrences of Oatly in company_links
links_violife = sum(1 for link in company_links_local_html if "violife" in link.lower())

# Print the result
print("Number of links for Violife:", links_violife)
```

```python
# Assuming company_links_local_html contains extracted URLs
company_links_local_html = [link["href"] for link in soup.find_all("a", class_="post-term-link")]

# Create an empty dictionary to count occurrences
url_counts = {}

# Count the number of times each URL appears
for url in company_links_local_html:
    if url in url_counts:
        url_counts[url] += 1
    else:
        url_counts[url] = 1

# Print the dictionary
print(url_counts)
```

```python
# Extract company names and construct URLs
base_url = "https://vegconomist.com/organisations-and-brands/"
company_names_map = {}  # Dictionary to map URLs to company names

for link in company_links_local_html:
    if base_url in link:
        company_name = link.replace(base_url, "").strip("/")
        company_names_map[link] = company_name  # Store mapping of URL to company name

# Replace URLs with company names in url_counts dictionary
articles_media_per_company = {company_names_map[url]: count for url, count in url_counts.items() if url in company_names_map}

# Sort dictionary alphabetically by company names
articles_media_per_company = dict(sorted(articles_media_per_company.items()))

# Print the updated dictionary
print(articles_media_per_company)
```

```python
import pandas as pd

# Convert dictionary to DataFrame
df = pd.DataFrame(list(articles_media_per_company.items()), columns=["company", "articles_media_per_company"])

# Print the DataFrame
print(df)
```

```python
# Function to clean company names with hypens
def clean_company_hyphens(name):
    if isinstance(name, str):
        name = name.lower()
        name = name.replace("  ", " ")
        name = name.replace(" ", "-")
        name = name.replace(",", "-")
        name = name.replace(".", "-") if not name.endswith(".") else name[:-1]
        name = name.replace("[", "-").replace("]", "-")
        name = name.replace("_", "-")
        name = name.replace("!", "-")
    return name

df_gfi["Company"] = df_gfi["Company"].apply(clean_company_hyphens)

print(df_gfi)
```

```python
import unicodedata

# Function to remove accent marks and umlauts
def remove_accents(name):
    if isinstance(name, str):
        name = unicodedata.normalize('NFKD', name).encode('ASCII', 'ignore').decode('utf-8')
    return name

df_gfi["Company"] = df_gfi["Company"].apply(remove_accents)

print(df_gfi)
```

```python
# Function to clean company names by replacing punctuation marks (orthographic)

def clean_company_orthographic(x):
    if isinstance(x, str):
        x = x.replace("&", "")
        x = x.replace("'", "")
        x = x.replace("(", "").replace(")", "")
        x = x.replace(":", "")
    return x

df_gfi["Company"] = df_gfi["Company"].apply(clean_company_orthographic)

print(df_gfi)
```

```python
import pandas as pd

# Function to check category presence
def check_cheese(product_types):
    return any("cheese" in p.lower() for p in product_types)

def check_other_dairy(product_types):
    return any("other dairy" in p.lower() for p in product_types)

# Apply transformations to merged_df
merged_df["Product Type"] = merged_df["Product Type"].fillna("").apply(lambda x: x.split(","))  # Splits values at commas

# Create new columns
merged_df["Cheese"] = merged_df["Product Type"].apply(check_cheese)
merged_df["Other dairy"] = merged_df["Product Type"].apply(check_other_dairy)
merged_df["Both cheese and other dairy"] = merged_df["Cheese"] & merged_df["Other dairy"]
merged_df["Neither cheese nor other dairy"] = ~(merged_df["Cheese"] | merged_df["Other dairy"])

# Convert boolean columns to integer (0 for False, 1 for True)
merged_df[["Cheese", "Other dairy", "Both cheese and other dairy", "Neither cheese nor other dairy"]] = merged_df[
    ["Cheese", "Other dairy", "Both cheese and other dairy", "Neither cheese nor other dairy"]
].astype(int)

# Display the modified DataFrame
print(merged_df[["Company", "Product Type", "Cheese", "Other dairy", "Both cheese and other dairy", "Neither cheese nor other dairy"]].head())
```

```python
# Filter dataframe based on conditions
filtered_df_product_range_dairy = merged_df_4_product_type_columns[
    (merged_df_4_product_type_columns["Protein category"].str.contains("Plant-based", case=False, na=False)) &
    (merged_df_4_product_type_columns["Technology Focus"].str.contains("End product formulation and manufacturing", case=False, na=False))
]

# Remove rows where all three columns ("Cheese", "Other dairy", "Both cheese and other dairy") are 0
filtered_df_product_range_dairy = filtered_df_product_range_dairy[(filtered_df_product_range_dairy["Cheese"] == 1) | (filtered_df_product_range_dairy["Other dairy"] == 1) | (filtered_df_product_range_dairy["Both cheese and other dairy"] == 1)]

# Create pivot table
pivot_table = filtered_df_product_range_dairy.pivot_table(
    values="Company",  # Counting companies (rows)
    index="HQ Country",  # Countries as row labels
    columns=["Cheese", "Other dairy", "Both cheese and other dairy"],  # Columns based on product types
    aggfunc="count",  # Count number of companies
    fill_value=0  # Fill missing values with zero
)

# Display the pivot table
print("Product range of plant-based dairy manufacturers in Europe:")
print(pivot_table)
```
